```python
def scrape_data(urls):
    data = []

    for url in urls:
        # Send a request to the URL and get the response
        response = requests.get(url)

        # Use BeautifulSoup to parse the HTML content of the page
        soup = BeautifulSoup(response.content, "html.parser")

        # Find the title and table data on the page
        title = soup.find("div", class_="section-title")
        ikan = title.find_all("p")[0].text.split('"')[1].replace("ikan ", "")
        pasar = title.find("p").get_text().split('Pasar "')[1].split('"')[0]
        table = soup.find("table")

        # Extract the data from the table and store it in a list of dictionaries
        for row in table.find_all("tr")[1:]:
            cols = row.find_all("td")
            harga_text = cols[2].text.strip().replace(".", "")  # Remove thousand separators
            harga_numerical = int("".join(filter(str.isdigit, harga_text)))  # Extract numerical part
            data.append({
                "Tanggal": cols[1].text.strip(),
                "Nama Ikan": ikan,
                "Nama Pasar": pasar,
                "Harga": harga_numerical,
            })

    # Convert the list of dictionaries to a Pandas DataFrame
    df = pd.DataFrame(data)
    return df

desired_df = scrape_data(urls)
```

```python
import requests
import pandas as pd
import numpy as np

url = 'https://datacenterpds.id/priceList'
response = requests.get(url)

json_data = response.json()
df = pd.json_normalize(json_data['data'])
df['harga'] = df['harga'].str.replace(',', '').astype(int)
df['tanggal_input'] = pd.to_datetime(df['tanggal_input'])
```

```python
df
```

```python
sliced_df = df.loc[df['nama_pasar'] == 'PASAR IKAN MODERN MUARA BARU']
sliced_df = sliced_df.loc[sliced_df['nama_ikan'].isin(['kembung\nlelaki', 'tongkol abu-abu', 'tenggiri', 'bandeng', 'tongkol\nkomo'])]
sliced_df['nama_ikan'] = sliced_df['nama_ikan'].str.replace('\n', ' ')
sliced_df.drop(['provinsi', 'kabupaten', 'nama_latin', 'nama_pasar'], axis=1, inplace=True)
sliced_df
```

```python
# Function to apply outlier filter using z-score method
def outlier_filter_zscore(df, column_name, threshold=3):
    z_scores = np.abs((df[column_name] - df[column_name].mean()) / df[column_name].std())
    filtered_df = df[z_scores <= threshold].copy()
    return filtered_df

# Apply outlier filter to DataFrame
filtered_df = outlier_filter_zscore(sliced_df, 'harga', threshold=3)
indexes = filtered_df[(filtered_df['harga'] < 10000)].index
filtered_df.drop(indexes , inplace=True)
```

```python
filtered_df.sort_values('tanggal_input', inplace=True)
filtered_df['tanggal_input'] = filtered_df['tanggal_input'].dt.date
filtered_df = filtered_df.groupby(['tanggal_input', 'nama_ikan'])['harga'].median().reset_index()
filtered_df['tanggal_input'] = pd.to_datetime(filtered_df['tanggal_input'])
```

```python
import requests
import pandas as pd
import numpy as np
from google.cloud import storage

def outlier_filter_zscore(df, column_name, threshold=3):
    z_scores = np.abs((df[column_name] - df[column_name].mean()) / df[column_name].std())
    filtered_df = df[z_scores <= threshold].copy()
    return filtered_df

def save_json_to_csv(request):
    url = 'https://datacenterpds.id/priceList'
    json_data = requests.get(url).json()
    df = pd.json_normalize(json_data['data'])
    df['harga'] = df['harga'].str.replace(',', '').astype(int)
    df['tanggal_input'] = pd.to_datetime(df['tanggal_input'])
    df = df.loc[df['nama_pasar'] == 'PASAR IKAN MODERN MUARA BARU']
    df = df.loc[df['nama_ikan'].isin(['kembung\nlelaki', 'tongkol abu-abu', 'tenggiri', 'bandeng', 'tongkol\nkomo'])]
    df['nama_ikan'] = df['nama_ikan'].str.replace('\n', ' ')
    df.drop(['provinsi', 'kabupaten', 'nama_latin', 'nama_pasar'], axis=1, inplace=True)
    df = outlier_filter_zscore(sliced_df, 'harga', threshold=3)
    indexes = df[(df['harga'] < 10000)].index
    df.drop(indexes , inplace=True)
    df.sort_values('tanggal_input', inplace=True)
    df['tanggal_input'] = df['tanggal_input'].dt.date
    df = df.groupby(['tanggal_input', 'nama_ikan'])['harga'].median().reset_index()
    df['tanggal_input'] = pd.to_datetime(df['tanggal_input'])
    client = storage.Client()
    bucket_name = 'my_personal_bucket_1'
    bucket = client.get_bucket(bucket_name)
    blob = bucket.blob('data_harga_pds.csv')
    blob.upload_from_string(df.to_csv(index=False), 'text/csv')

    return 'Data saved successfully!'
```
